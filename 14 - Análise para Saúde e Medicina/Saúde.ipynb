```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

display(deaths.head())
display(recovered.head())
```

```python
def latest_by_country(data):
    return data.groupby("Country/Region").sum().iloc[:,-1]

latest_by_country(confirmed)
```

```python
info = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(info, axis = 1)
combined.columns = ["Confirmados", "Mortes", "Recuperados"]
combined
```

```python
# Taxa de letalidade
sum_up = combined.sum()
letality_rate_1 = sum_up["Mortes"]/ sum_up["Confirmados"]
letality_rate_1 * 100
```

```python
letality_per_country = combined["Mortes"]/ combined["Confirmados"] * 100
combined["Letalidade por país (%)"] = letality_per_country
combined
```

```python
letality_rate_2 = combined["Mortes"]/ (combined["Mortes"] + combined["Recuperados"])
letality_rate_2 *= 100
letality_rate_2
```

```python
combined.sort_values("Letalidade por país (%)", ascending = False).head(10)
```
